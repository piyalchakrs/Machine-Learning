```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('emails.csv')
```

```python
df
```

```python
df['spam'].value_counts()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

# sperate x and y

```python
x = df.text.values
```

```python
y = df.spam.values
```

# split dataset

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

# Data preprocessing

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_train = cv.fit_transform(xtrain)
```

```python
x_train.toarray()
```

# ML Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(x_train,ytrain)
```

```python
x_test = cv.transform(xtest)
```

```python
x_test.toarray()
```

```python
model.score(x_test,ytest)
```

```python
emails = ['Hey i am looking for machine learning tutorial in bengali language',
           'hey you won an iphone x giveaway for free please do the suvey ']
```

```python
cv_emails = cv.transform(emails)
```
