```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Car_data.csv')
df.head()
```

```python
df.driver_experience
```

# Handling the null value

```python
df.driver_experience = df.driver_experience.fillna(df.driver_experience.mean())
df.head()
```

```python
df.car_age = df.car_age.fillna(df.car_age.mean())
```

```python
df
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
X = df[['speed', 'car_age', 'driver_experience']]
y = df['risk']
```

```python
reg.fit(X,y)
```

```python
reg.predict([[160,10,5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X,y)
```
